# 运算

## 一.索引与切片

(1)基本索引

```python
import numpy as np
arr=np.random.randint(1,100,10)#若10替换成(1,10)就变成了矩阵
print(arr[1])#引用第二个
print(arr[-1])#倒数第一个
print(arr[:])#引用全部
print(arr[2:9])#第三个到第九个（因为左闭右开）
#等价的可以用slice函数
print(arr[slice(2,5)])
print(arr[slice(2,10,2)])#加了个步长
#bool索引（满足条件的被索引）支持逻辑运算符(and& or| not~)
print(arr[arr>10])
print(arr[(arr>10)&(arr<50)])
print(arr[~(arr<50)])
```

矩阵的索引

```python
import numpy as np
arr=np.random.randint(1,100,(10,10))
print(arr)
print(arr[2])#引用第三行
print(arr[2,2:7])#引用第三行第3个数到第七个数
print(arr[2,6])#第三行第七列
#用slice函数
print(arr[slice(2)])
#bool索引
print(arr[arr>30])
print(arr[3][arr[3]>40])#第四行的bool索引
```

3.常用函数

基本数学

```python
import numpy as np
p=4
q=9
a1=np.sqrt(p)#平方
a2=np.exp(p)
a3=np.log(p)#以e为底
a4=np.sin(p)
a5=np.power(p,q)#a的b次幂
a6=np.round(q)#向下取整，四舍五入
print(a1,a2,a3,a4,a5,a6)
```

统计

```python
import numpy as np
x=np.random.randint(1,10,(1,10))
y=np.random.randint(1,10,(1,10))
a1=np.sum(x)#求和
a2=np.mean(y)#平均数
a3=np.median(y)#中位数
a4=np.std(x)#标准差
a5=np.var(x)#方差
a6=np.min(x)
a7=np.max(x)
a8=np.percentile(y,40)#百分之40分位数
print(a1,a2,a3,a4)
print(a5,a6,a7,a8)
print(np.sum(x))
print(np.cumprod(x))#累计乘积，第i个 和前i-1个乘
```

比较


```python
import numpy as np
x=0
y=9
a1=np.greater(x,y)#x是否比y大
a2=np.less(x,y)
a3=np.logical_and(x,y)#逻辑运算符
a4=np.logical_or(x,y)
z=np.random.randint(1,10,(1,10))

a5=np.logical_not(z)
a6=np.logical_not(x)
print(a1,a2)
print(z)
print(a3)
print(a4)
print(a6)
print(a5)
print(np.where(z>3,z,0))#符合条件从z里取 不符合变成0
print(np.where(z>3,1,0))#符合条件变成1 不符合变成0


```

```python
#检测考试是否及格
import numpy as np
score=np.random.randint(30,100,20)
print(score)
print(np.where(score<60,'不及格',np.where(score<80,'良好','优秀')))
```

```python
#np.select条件
import numpy as np
score=np.random.randint(30,100,20)
print(score)
print(np.select([score<60,(score>=60)&(score<=80),score>80],['不及格','良好','优秀'],default='未知'))
```

排序

```python
import numpy as np
np.random.seed(125)
arr=np.random.randint(1,300,30)
print(arr)
#arr.sort()
#print(arr)这样会改原始数据 污染了原始数据 但是下边的就不会 见下
print(np.sort(arr))
print(np.argsort(arr))#排序后的索引
print(arr)
```
